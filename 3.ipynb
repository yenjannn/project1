## 3-1 這是哪種鳶尾花

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
x = iris.data[:, :2]
y = iris.target
```

# SVC

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x, y)
```

```python
clf.predict(x)
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x))
```

# K-Mean

```python
from sklearn.cluster import KMeans
```

```python
clfk = KMeans(n_clusters=3)
```

```python
clfk.fit(x)
```

```python
clfk.labels_
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clfk.labels_)
```

## 3-4 圖形化我們的成果

鳶尾花的例子中，色塊是學習出來的，點是原始資料，是不是一下就看出哪些錯了?

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

上次的成果拿回來使用，記得我們做了個鳶尾花分類器。

找回分類器

```python
from sklearn.externals import joblib
clf = joblib.load("iris_clf_01.pkl")
```

```python
clf.predict([[2,3]])
```

看看分類的全貌

我們用一下之前的方式, 畫出我們想要看到我們可愛的 SVM 是怎麼以花萼長度、花萼寬度來分類的。上次我們用了 Python 所謂 "list comprehension" 的作法 (本質上是 for 迴圈), 現在我們換個方式看來比較「高級」的方式。

```python
xt, yt = np.meshgrid(np.arange(-2,2,0.5), np.arange(-1,1,0.5))
```

```python
xt
```

```python
yt
```

看得出來 meshgrid 做了什麼呢? 基本上它就是說我們在 x, y 兩個指定範圍的長方型當中, 依我們指定的間隔找出格點。這些格點的座標分成 x 座標一個 array, y 座標一個。x 或 y 座標的 array, 的座標是一列一列標記的。

要是你覺得這樣的表示法很討厭, 我們也可以讓它變一長串的向量。

```python
xt.ravel()
```

```python
xt  #注意這其實原來的 xt 並沒有變哦。
```

把 (x,y) 一點一點的座標收集起來。

```python
np.c_[xt.ravel(), yt.ravel()]
```

把資料的型式這樣變來變去會是數據分析非常非常常做的事情。

```python
xx, yy = np.meshgrid(np.arange(3, 8.5, 0.2), np.arange(1.5, 5.0, 0.2))
```

```python
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
plt.scatter(xx.ravel(), yy.ravel(), s=50, c=Z)
```

快速換個配色。

```python
plt.scatter(xx.ravel(), yy.ravel(), s=50, c=Z, cmap=plt.cm.coolwarm, alpha=0.6)
```

```python
plt.scatter(xx.ravel(), yy.ravel(), s=50, c=Z, cmap=plt.cm.prism, alpha=0.6)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
x = iris.data[:, :2]
y = iris.target
```

我們來畫畫比較。

```python
plt.subplot(121)
plt.scatter(x[:,0], x[:,1], s=50, c=y)

plt.subplot(122)
plt.scatter(x[:,0], x[:,1], s=50, c=clf.predict(x))
```

```python
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
```
