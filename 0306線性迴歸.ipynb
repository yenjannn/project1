```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# 線性迴歸

```python
x = np.linspace(0 , 5 , 50)
```

```python
y = 1.2*x + 0.8
```

```python
plt.scatter(x,y)
plt.plot(x, 1.2*x + 0.8, 'r')
```

```python
y = 1.2*x + 0.8 + 0.6*np.random.randn(50)
```

```python
plt.scatter(x,y)
plt.plot(x,1.2*x + 0.8,'r')
```

```python
from scipy.optimize import curve_fit
```

```python
def f(x, a, b):
    return a*x + b
```

```python
popt, pcov = curve_fit(f, x, y)
```

```python
popt
```

```python
a, b = popt
```

```python
a
```

```python
plt.scatter(x, y)
plt.plot(x, a*x + b, 'r')
```

# 不是線性的目標函數

```python
x = np.linspace(0 , 5 , 50)
```

```python
y = np.sin(5*x) + 3
```

```python
y = np.sin(5*x) + 3 + 0.2*np.random.randn(50)
```

```python
plt.scatter(x,y)
plt.plot(x, np.sin(5*x) + 3, 'r')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state=1)
```

```python
plt.scatter(x_train, y_train)
```

```python
def g(x, a, b):
    return np.sin(a*x) + b
```

```python
popt, pcov = curve_fit(g, x_train, y_train, p0=[4,2]) #猜起始值
```

```python
a, b = popt
```

```python
x_trainsorted = np.sort(x_train)
```

```python
x_trainsorted
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_trainsorted, np.sin(a*x_trainsorted) + b,'r')
```
