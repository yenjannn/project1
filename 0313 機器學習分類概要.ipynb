# 機器學習分類概要

* supervised learning   監督→有告訴正確答案
* unsupervised learning 非監督→不提供正確答案

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[-3, 2],[-6, 5],[3, -4],[2, -8]])
y = np.array([1, 1, 2, 2])
```

```python
x[:]
```

```python
x[:,0]
```

```python
plt.scatter(x[:,0], x[:,1], s=50,c=y )
```

# SVM  -supervised

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x, y)
```

```python
clf.predict([[-0.8, 1]])
```

```python
gd = np.array([[i, j] for i in np.arange(-8, 4, 0.6) 
              for j in np.arange(-10, 5, 0.8)]) 
```

```python
gd
```

```python
clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=clf.predict(gd))
```

## 2

```python
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=3)
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

```python
clf2 = SVC()
```

```python
clf2.fit(x, y)
```

```python
plt.subplot(211)
plt.scatter(x[:,0], x[:,1], s=50, c=y)
plt.subplot(212)
plt.scatter(x[:,0], x[:,1], s=50, c=clf2.predict(x))
```

```python
gd = np.array([[i, j] for i in np.arange(-3, 6, 0.4) 
              for j in np.arange(-3, 4, 0.4)]) 
```

```python
gdc = clf2.predict(gd)
```

```python
clf2.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```

# K-mean -unsupervised

```python
x = np.random.randn(100, 2)
```

```python
x
```

```python
plt.scatter(x[:,0], x[:,1], s=50)
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0], x[:,1], s=50, c=clf.labels_)
```

```python
gd = np.array([[i, j] for i in np.arange(-3, 3, 0.4) 
              for j in np.arange(-3, 3, 0.4)]) 
```

```python
gdc = clf.predict(gd)
```

```python
plt.scatter(gd[:,0], gd[:,1], s=50, c=gdc)
```
